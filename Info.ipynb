## Stocker une image dans une BD

### **1. Stockage direct dans la base (BLOB / bytea)**


* **Principe** : l’image est convertie en flux binaire (octets) et stockée directement dans une colonne de type `BLOB` (Binary Large Object) ou équivalent (`bytea` en PostgreSQL, `VARBINARY(MAX)` en SQL Server, `LONG BLOB` en MySQL, etc.).
* **Avantages** :

  * Tout est centralisé dans la base (données et fichiers).
  * Sauvegarde/restauration simplifiée (tout est inclus dans les dumps).
* **Inconvénients** :

  * La base peut devenir volumineuse et lente.
  * Moins efficace que le système de fichiers pour des images très lourdes ou en grand nombre.

### **2. Stockage encodé en texte (Base64)**


* **Principe** : l’image est encodée en Base64 et stockée dans un champ texte (`TEXT`, `CLOB`, etc.).
* **Avantages** :

  * Compatible avec des environnements qui ne gèrent pas bien le binaire.
  * Facile à transporter via JSON, XML ou API.
* **Inconvénients** :

  * L’encodage Base64 augmente la taille du fichier d’environ 30%.
  * Performance moins bonne que les BLOB.

### **3. Stockage indirect (référence / chemin)**

* **Principe** : on ne stocke pas l’image dans la base, mais uniquement son chemin d’accès ou son URL (par exemple `/images/produits/123.jpg` ou `https://cdn.monsite.com/img/123.jpg`).
* **Avantages** :

  * Très performant, car les images restent dans le système de fichiers ou un service de stockage (cloud : AWS S3, Google Cloud Storage, Azure Blob Storage).
  * La base reste légère et rapide.
* **Inconvénients** :

  * Les sauvegardes doivent inclure à la fois la base et les fichiers/stockages externes.
  * Gestion plus complexe de la cohérence (ex. : entrée en base mais fichier manquant).

### **4. Stockage hybride**

* **Principe** : stocker les métadonnées et une miniature (thumbnail) en base, mais les images originales dans un système de fichiers ou un stockage cloud.
* **Avantages** :

  * Combine performance (base légère) et praticité (aperçus rapides).
* **Inconvénients** :

  * Nécessite une logique applicative plus élaborée.

✅ **En pratique** :

* **Applications simples ou petites images** → BLOB ou Base64.
* **Applications web/mobiles avec beaucoup d’images** → stockage sur disque/Cloud + URL en base (solution la plus répandue).
* **Cas mixtes (galeries, e-commerce)** → stockage hybride.
